```python
import heapq

def prim_mst(graph, start):
    visited = set()
    min_heap = [(0, start)]   # (weight, vertex)
    
    mst = []
    total_weight = 0

    while min_heap:
        weight, u = heapq.heappop(min_heap)

        if u in visited:
            continue

        visited.add(u)
        total_weight += weight

        if weight != 0:
            mst.append((u, weight))

        for v, w in graph[u]:
            if v not in visited:
                heapq.heappush(min_heap, (w, v))

    return mst, total_weight


graph = {
    0: [(1,4),(7,8)],
    1: [(0,4),(2,8),(7,11)],
    2: [(1,8),(3,7),(8,2),(5,4)],
    3: [(2,7),(4,9),(5,14)],
    4: [(3,9),(5,10)],
    5: [(2,4),(3,14),(4,10),(6,2)],
    6: [(5,2),(7,1),(8,6)],
    7: [(0,8),(1,11),(6,1),(8,7)],
    8: [(2,2),(6,6),(7,7)]
}

mst, total_weight = prim_mst(graph, 0)

print("Edges in MST:")
for edge in mst:
    print(edge)

print("Total Weight:", total_weight)
```
