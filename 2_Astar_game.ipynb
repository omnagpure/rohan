```python
import heapq

goal = [[1,2,3],
        [4,5,6],
        [7,8,0]]

def h(state):
    count = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0 and state[i][j] != goal[i][j]:
                count += 1
    return count

def find0(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

def print_state(state):
    for row in state:
        print(row)
    print()

def astar(start):

    pq = []
    heapq.heappush(pq, (h(start), start))

    visited = []

    while pq:

        cost, state = heapq.heappop(pq)

        print_state(state)

        if state == goal:
            print("Goal Reached")
            return

        visited.append(state)

        x, y = find0(state)

        moves = [(-1,0),(1,0),(0,-1),(0,1)]

        for dx, dy in moves:

            nx, ny = x + dx, y + dy

            if 0 <= nx < 3 and 0 <= ny < 3:

                new = [row[:] for row in state]

                new[x][y], new[nx][ny] = new[nx][ny], new[x][y]

                if new not in visited:
                    heapq.heappush(pq, (h(new), new))

start = [[1,2,3],
         [4,0,6],
         [7,5,8]]

astar(start)
```
