```python
graph = [
    (0,1,4), (0,7,8), (1,2,8), (1,7,11),
    (2,3,7), (2,8,2), (2,5,4), (3,4,9),
    (3,5,14), (4,5,10), (5,6,2),
    (6,7,1), (6,8,6), (7,8,7)
]

parent = [i for i in range(9)]

def find(x):

    while parent[x] != x:
        x = parent[x]

    return x

cost = 0
graph.sort(key=lambda x: x[2])
print("Edges in MST:")

for u, v, w in graph:
    x = find(u)
    y = find(v)

    if x != y:
        parent[x] = y
        print((u, v, w))
        cost += w

print("\nTotal Cost =", cost)
```
