```python
print("Welcome to RedBus Chatbot")
print("Type 'exit' to stop\n")

while True:

    user = input("You: ").lower()

    if user == "hi" or user == "hello":
        print("Bot: Hello! Welcome to RedBus")

    elif "book" in user:
        print("Bot: Please enter source and destination")

    elif "ticket" in user:
        print("Bot: Your ticket has been booked successfully")

    elif "cancel" in user:
        print("Bot: Your ticket cancellation request is processed")

    elif "price" in user or "fare" in user:
        print("Bot: Ticket price starts from Rs.500")

    elif "bus" in user:
        print("Bot: Available buses are Sleeper, AC, and Volvo")

    elif user == "exit":
        print("Bot: Thank you for using RedBus Chatbot")
        break

    else:
        print("Bot: Sorry, I did not understand")
```
