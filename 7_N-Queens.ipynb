```python
def isSafe(row,col,board):
  for i in range(row):
    
    if board[i]==col:
      return False
   
    if abs(board[i]-col)==abs(i-row): return False
  return True
```

```python
def solve_nQueens(n):
  board=[-1]*n
  sol=[]

  def solve(row):
    if row==n:
      sol.append(board.copy())
      return

    for c in range(n):
      if isSafe(row,c,board):
        
        board[row]=c
        solve(row+1)
        board[row]=-1

  solve(0)
  return sol
```

```python
def printBoard(sol,n):
  for i in range(n):
    row=['.']*n
    row[sol[i]]='Q'
    print(" ".join(row))
  print()
```

```python
n=int(input("Enter value of N: "))
sol=solve_nQueens(n)

print("\nTotal Solutions: ",len(sol))
for i, sol in enumerate(sol,1):
  print(f"\nSolution {i}: {sol}")
  printBoard(sol,n)
```
