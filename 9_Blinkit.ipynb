```python
def blinkit_chatbot():
    print("🤖 Welcome to Blinkit Customer Chatbot 🛒")
    print("I can help you with orders, delivery, refund, and items.")
    print("Type 'exit' to stop.\n")

    while True:
        user = input("You: ").lower()

        if user == "exit":
            print("Bot: Thank you for using Blinkit. Happy Shopping 🛍️")
            break

           # Help menu
        elif user == "help":
            print("Bot: I can help you with:")
            print("- booking bus ticket")
            print("- cancellation")
            print("- refund")
            print("- timing")
            print("- routes")

        elif "order" in user or "buy" in user:
            print("Bot: You can place an order on Blinkit app by selecting items and confirming checkout.")

        elif "delivery" in user or "time" in user:
            print("Bot: Blinkit delivers groceries within 10–30 minutes depending on your location.")

        elif "refund" in user or "return" in user:
            print("Bot: Refund will be processed within 2–5 working days after return approval.")

        elif "cancel" in user:
            print("Bot: You can cancel your order before it is packed from 'My Orders' section.")

        elif "items" in user or "product" in user:
            print("Bot: Blinkit offers groceries, fruits, vegetables, dairy, snacks, and daily essentials.")

        elif "help" in user:
            print("Bot: I can help you with orders, delivery time, refund, cancellation, and products.")

        else:
            print("Bot: Sorry, I didn't understand. Type 'help' for options.")


# Run chatbot
blinkit_chatbot()


```
