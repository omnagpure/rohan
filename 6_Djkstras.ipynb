```python
import heapq

graph = {
    'A': [('B',4), ('C',5)],
    'B': [('D',9), ('E',7)],
    'C': [('E',3)],
    'D': [('F',2)],
    'E': [('F',6)],
    'F': []
}

pq = [(0, 'A')]
visited = {}

while pq:
    cost, node = heapq.heappop(pq)

    if node not in visited:
        visited[node] = cost

        for n, w in graph[node]:
            heapq.heappush(pq, (cost + w, n))

print("Shortest Distance from A:")

for i in visited:
    print(i, "=", visited[i])
```
