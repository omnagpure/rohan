```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# First time only:
# nltk.download('punkt')
# nltk.download('wordnet')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def ifb_chatbot():
    print("🤖 Welcome Dear Customer to IFB Service Centre 🧺")
    print("Ask me about service, repair, booking, warranty, complaint\n")

    while True:
        user = input("You: ").lower()

        if user == "exit":
            print("Bot: Thank you for contacting IFB Service Centre 🙏")
            break

        # 🔹 Tokenization
        words = word_tokenize(user)

        # 🔹 Stemming
        stem_words = [stemmer.stem(w) for w in words]

        # 🔹 Lemmatization
        lemma_words = [lemmatizer.lemmatize(w) for w in words]

        # (Optional - for viva demonstration)
        print("Tokens:", words)
        print("Stemmed:", stem_words)
        print("Lemmatized:", lemma_words)

        # 🔹 Intent detection (use lemma_words for better meaning)

        if "service" in lemma_words or "repair" in lemma_words:
            print("Bot: You can book a service request via IFB website or helpline.")

        elif "book" in lemma_words or "booking" in lemma_words or "schedule" in lemma_words:
            print("Bot: Service booking is available online. Choose date and time.")

        elif "warranty" in lemma_words:
            print("Bot: IFB provides standard warranty. Please check invoice details.")

        elif "complaint" in lemma_words or "issue" in lemma_words:
            print("Bot: You can register complaints via IFB service portal.")

        elif "cost" in lemma_words or "price" in lemma_words:
            print("Bot: Service cost depends on issue. Technician will inform you.")

        elif "help" in lemma_words:
            print("Bot: I can help with service, booking, repair, warranty, complaints.")

        else:
            print("Bot: Sorry, I didn't understand. Please try again or type 'help'.")


# Run chatbot
ifb_chatbot()
```
