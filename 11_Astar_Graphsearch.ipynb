```python
graph = {
    'a': [('b', 4), ('c', 3)],
    'b': [('f', 5), ('e', 12)],
    'c': [('e', 10), ('d', 7)],
    'd': [('e', 2)],
    'e': [('z', 5)],
    'f': [('z', 16)],
    'z': []
}

heuristic = {
    'a': 14,
    'b': 12,
    'c': 11,
    'd': 6,
    'e': 4,
    'f': 11,
    'z': 0
}

def astar(start, goal):

    open_list = [(start, 0)]
    visited = []

    while open_list:

        open_list.sort(key=lambda x: x[1])

        current, cost = open_list.pop(0)

        print(current, end=" ")

        if current == goal:
            print("\nGoal Reached")
            return

        visited.append(current)

        for neighbour, edge_cost in graph[current]:

            if neighbour not in visited:

                f = edge_cost + heuristic[neighbour]

                open_list.append((neighbour, f))


print("A* Traversal:")
astar('a', 'z')
```
