```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['G'],
    'F': [],
    'G': []
}

heuristic = {
    'A': 5,
    'B': 4,
    'C': 3,
    'D': 6,
    'E': 2,
    'F': 4,
    'G': 0
}

def greedy(start, goal):

    open_list = [start]
    visited = []

    while open_list:
        open_list.sort(key=lambda x: heuristic[x])
        current = open_list.pop(0)

        print(current, end=" ")

        if current == goal:
            print("\nGoal Reached")
            return

        visited.append(current)

        for neighbour in graph[current]:

            if neighbour not in visited:
                open_list.append(neighbour)

print("Greedy Search Traversal:")
greedy('A', 'G')
```
